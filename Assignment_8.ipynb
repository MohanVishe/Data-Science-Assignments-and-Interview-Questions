```python
# Return statement is used to return the output from the fuction
# The fuction can have multiple return statements
# They can be seperated by conditions and any return statement run the function will exit
# The return statement also can have multiple values which are seperated by comma
```

```python
def prime(num):
  """Give Sring which tell prime or not"""
  if num==1:
    return "Not prime nor composite"
  a=True
  for i in range(2,num):
    if num%i==0:
      a=False

  if a:
    return "Prime"
  else:
    return "Not Prime"
```

### <font color='red'>7. What is the difference between the "pass by value" and "pass by reference" concepts when it comes to function arguments in Python?</font>

```python
""" Pass by value """
# Any other operation performed in function not have any effect on the original value
# Because the argument passed to function is copied.
# It changes the value of copy created inside function

def sq(a):
  a=a**2
  print("inside",a)


a=100
print("before",a)
sq(a)
print("after",a)
```

```python
"""Pass by reference"""
# Any other operation performed in the function wil affect the original value
# Because arguments passed are reference of the orginal object
# It changes value inside and outside of the function

def add(a):
  a.append(10)
  print("inside",a)


a=[2,34,5,3]
print("before",a)
add(a)
print("after",a)
```
