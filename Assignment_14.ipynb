- **Answer**



```sql
create database Ecoweb;
use Ecoweb;

-- create table product info
create table pro_info(
pr_id int ,
primary key(pr_id),
product varchar(20) not null );

-- create table likes
create table likes (
us_id int ,
pr_id int ,
foreign key (pr_id) references pro_info(pr_id),
like_date date
);

-- Inserting values
insert into pro_info (pr_id,product)
values (1001,"Blog"),(1002,"Youtube"),(1003,"Education");

insert into likes(us_id,pr_id,like_date)
values (1,1001,'2023-06-27'),(2,1003,'2023-06-23');

-- Solution

select distinct pi.pr_id from pro_info as pi  left outer join likes as l
on pi.pr_id=l.pr_id  where l.like_date is Null;
```
